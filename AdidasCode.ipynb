```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

adidas_DF = pd.read_csv("C:/Users/user/Desktop/Adidas sales/adidas.csv")
adidas_DF.head(10)
```

```python
print(adidas_DF.columns) # columns
df = adidas_DF.copy()
df.sample(10)
```

```python
# Profit by retailer
retailer_profit = df.groupby('Retailer')['Operating Profit'].sum()
retailer_profit = retailer_profit.sort_values(ascending=True)
print('profit by retailer')

print(retailer_profit)
most_profitable_retailer = retailer_profit.idxmax()
highest_profit = retailer_profit.max()

print(f'The most profitable retailer is {most_profitable_retailer} with a total profit of {highest_profit}')
# bar plot
plt.figure(figsize=(8, 5))
sns.set_style('darkgrid')
plt.rcParams.update({'font.size' :8})
sns.barplot(retailer_profit, width=0.5)
plt.title('Profit by retailer', weight='bold', fontsize=16)
plt.show()
```

```python
# The most efficient retailer in generating profit based on the operating margin
print('most efficient retailer in profit generation')

retailer_margin = df.groupby('Retailer')['Operating Margin'].mean()
retailer_margin = retailer_margin.sort_values(ascending=True)
print(retailer_margin)
most_efficient_retailer = retailer_margin.idxmax()
highest_margin = retailer_margin.max()

print(f'The most efficient retailer in generating profit is {most_efficient_retailer} with an average operating margin of {highest_margin}')

# bar plot
plt.figure(figsize=(8, 5))
sns.set_style('darkgrid')
plt.rcParams.update({'font.size' :8})
sns.barplot(retailer_margin, width=0.5)
plt.title('Most efficient retailer in profit generation', weight='bold', fontsize=16)
plt.show()
```

```python
# Sales trends over time
monthly_trends = df.groupby(df['Invoice Date'].dt.month_name()).agg({'Units Sold': 'sum',
                                                                           'Total Sales': 'sum',
                                                                           'Operating Profit': 'sum'})
Order = ['January', 'February', 'March', 'April', 'May', 'June',
       'July', 'August', 'September','October', 'November', 'December']

monthly_trends.index = pd.Categorical(monthly_trends.index, categories=Order, ordered=True)
monthly_trends.index.name = 'month'
monthly_trends.sort_values(by='month')
print(monthly_trends)

# line plot
plt.figure(figsize=(8, 4))
sns.set_style('darkgrid')
plt.rcParams.update({'font.size' :8})
sns.lineplot(monthly_trends.iloc[:, 0:1])
plt.title('Sales trends over time', weight='bold', fontsize=16)
plt.show()

# total sales and operating profit
plt.figure(figsize=(8, 4))
sns.set_style('darkgrid')
plt.rcParams.update({'font.size' :8})
sns.lineplot(monthly_trends.iloc[:, 1:])
plt.title('Sales trends over time', weight='bold', fontsize=16)
plt.show()
```

```python
# Profit by rregion
regional_profit = df.groupby('Region').agg({'Units Sold': 'sum',
                                            'Total Sales': 'sum',
                                            'Operating Profit': 'sum'})
#regional_profit = retailer_profit.sort_values(by='Operating Profit', ascending=True)
print('profit by region')

print(regional_profit)
most_profitable_region = regional_profit['Operating Profit'].idxmax()
highest_nregio = regional_profit['Operating Profit'].max()
print(f'The most profitable region is {most_profitable_region} with a total profit of {highest_profit}')

most_units_sold_region = regional_profit['Units Sold'].idxmax()
most_units_sold = regional_profit['Units Sold'].max()
print(f'region with the most units sold is {most_units_sold_region} selling a total of {most_units_sold}')


# bar plot
plt.figure(figsize=(8, 5))
sns.set_style('darkgrid')
plt.rcParams.update({'font.size' :8})
sns.barplot(regional_profit.iloc[:, 0:1],x=regional_profit.index, y='Units Sold', width=0.5)
plt.title('Units sold by region', weight='bold', fontsize=16)
plt.show()

sns.barplot(regional_profit.iloc[:, 1:],x=regional_profit.index, y='Total Sales', width=0.5)
plt.title('Profit by region', weight='bold', fontsize=16)
plt.show()

sns.barplot(regional_profit.iloc[:, 1:],x=regional_profit.index, y='Operating Profit', width=0.5)
plt.title('Profit by region', weight='bold', fontsize=16)
```
